```python
# load data
import pandas as pd
import numpy as np

# visualization
import matplotlib.pyplot as plt

# modelling 
import statsmodels.formula.api as smf

# statistics 
import scipy.stats as stats
```

```python
# read data Kindle
kindle_data = pd.read_csv("Kindle_Data_Cleaned_2.csv")

# memastikan tiap kolom tidak mengandung spasi yang tak terlihat
kindle_data.columns = kindle_data.columns.str.strip()

# menampilkan semua data Books
kindle_data.head(3)
```

```python
# menambah kolom sales yaitu perkalian antara reviews dan price
kindle_data["sales"] = kindle_data["price"] * kindle_data["reviews"]

kindle_data.head(3)
```

```python
def print_coef_std_err(results):
    """
    Function to combine estimated coefficients and standard error in one DataFrame
    :param results: <statsmodels RegressionResultsWrapper> OLS regression results from 
    :return df: <pandas DataFrame> 
    """
    coef = results.params
    std_err = results.bse
    
    df = pd.DataFrame(data = np.transpose([coef, std_err]), 
                      index = coef.index, 
                      columns=["coef","std err"])
    return df
```

```python
# Create OLS model object
model = smf.ols("log_sales ~ stars", kindle_no_0_reviews)

# Fit the model
results = model.fit()

# Extract the results (Coefficient and Standard Error) to DataFrame
results_2 = print_coef_std_err(results)
results.summary()
```

```python
predictor = "stars"
outcome = "log_sales"
data = kindle_no_0_reviews.copy()
results_ = results_2.copy()

# Plot the data
plt.scatter(data[predictor], data[outcome], color = "k", marker=".")

# Calculate the fitted values
a_hat = results_2.loc["Intercept"]["coef"]
b_hat = results_2.loc[predictor]["coef"]

x_domain = np.linspace(np.min(data[predictor]), np.max(data[predictor]), 10000)

fitted_values = a_hat + b_hat * x_domain

# Plot the fitted line
plt.plot(x_domain, fitted_values, label="Fitted line", color = "b")

# Add a legend and labels
plt.legend()
plt.ylabel("log_sales")
plt.xlabel("stars")

# Add a title and adjust the margins
plt.title(f"Data and fitted regression line \n {outcome} = {np.round(a_hat,2)} + {np.round(b_hat,2)}{predictor}")

# Show the plot
plt.show()
```

```python
# Use LabelEncoder to convert the smoker variable into numeric
from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder Object and transform the smoker variable
kindle_no_0_reviews.loc[:,"isEditorsPick"] = LabelEncoder().fit_transform(kindle_no_0_reviews["isEditorsPick"])

# Display the 5th first row after transforming
kindle_no_0_reviews[["isEditorsPick","log_sales"]].head()
```

```python
# Create OLS model object
model = smf.ols('sales ~ stars + isEditorsPick', kindle_no_0_reviews)

# Fit the model
results = model.fit()

# Extract the results (Coefficient and Standard Error) to DataFrame
results_stars_editor = print_coef_std_err(results)
results.summary()
```

```python
predictors = ["stars","isEditorsPick"]
outcome = "sales"
data = kindle_no_0_reviews.copy()
results_ = results_stars_editor.copy()
linestyles = {0 : "--", 1: "-"}
c = {0 : "red", 1: "green"}
c_line = {0 : "r", 1: "g"}
markers = {0 : ".", 1: "+"}

# Plot the data
for i in range(2):
    plt.scatter(data[data[predictors[1]]==i][predictors[0]], 
                data[data[predictors[1]]==i][outcome], 
                color = c[i], 
                marker = markers[i])
    
x_domain = np.linspace(np.min(data[predictors[0]]), np.max(data[predictors[0]]), 100)


# Calculate the fitted values for two lines regarding the mom_hs
beta0_hat = results_.loc["Intercept"]["coef"]
beta1_hat = results_.loc[predictors[0]]["coef"]
beta2_hat = results_.loc[predictors[1]]["coef"]

fitted_values = [beta0_hat +  beta1_hat * x_domain  + beta2_hat * i for i in range(2)]


# Plot two fitted line
for i in range(2):
    plt.plot(x_domain, fitted_values[i], c = c_line[i], label = f"Fitted line (isEditorsPick={i})", linestyle = linestyles[i])

# Add a legend and labels
plt.legend()
plt.ylabel(f"{outcome}")
plt.xlabel(f"{predictors[0]}")

# Add a title and adjust the margins
plt.title("Data and fitted regression line")

# Show the plot
plt.show()
```

```python
def group_rating(rating):
    if rating >= 3.5:
        return "High Rating"
    elif rating >= 3:
        return "Medium Rating"
    else:
        return "Low Rating"

# Terapkan ke kolom stars
kindle_data["rating_group"] = kindle_data["stars"].apply(group_rating)
```

```python
# Mapping sesuai urutan yang diinginkan
rating_map = {
    'Low Rating': 0,
    'Medium Rating': 1,
    'High Rating': 2
}

# Terapkan mapping ke kolom rating_group
kindle_data["rating_group"] = kindle_data["rating_group"].map(rating_map)

# Cek hasilnya
kindle_data[["rating_group", "sales"]].head()
```
